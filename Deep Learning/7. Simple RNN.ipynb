#### Can we solve problem where sequence of data is very important with ANN ?


- **No**, we can't directly use ANN to solve the problem where sequence of data is very important as there are some drwabacks in ANN:
    - it is not able to maintain the entire sequence (meaning of the sentence is loss)
    - it will give all the word to ANN all at once for a specific sentence



To solve the issue and for solving the problem having sequence of data importance, we will be using **"Simple RNN"**

## Simple RNN - Recurrent Neural Network


- The main difference in the ANN and RNN Architecture is that RNN will having a feedback loop in all the neurons present in the hidden layers and it will send the feedback to itself and all the other neurons present in the hidden layers
- after processing the input along with weights, bias and activation function then whatever outpu we are getting will share it with all the other neurons present in the hidden layers



- in the input sentence here will be doing some numbering based on sentences and words, and pass the input at each timestamp means we pass one word at a time in RNN
- here the previous context is also retained by the RNN

### Working of Simple RNN with Forward Propogation


- Suppose we have applied OneHotEncoding to convert the word into vector
- for each timestamp will pass one word at a time to RNN 
- in the input layer, weights get applied, bias and activation function gets added along with self feedback loop whose output will be going to passing to each neurons present in the hidden layer
- so the context of previous word will also be present in each case 



- suppose in RNN, 5 inputs are there, 3 neurons are there in the hidden layer and 1 neuron in output layer then:
    - weights from input layer: 5x3 = 15
    - weights in hidden layer (self loop): (1x5).(5x3) = 3x3 = 9
    - weights from hidden layer: 3
    - 3 bias in hidden layer and 1 bias in output
    - so Total Trainable Parameters = 31

### Problems with RNN
- As we know, ANN will face problem of Vanishing Gradient Problem so similar problem we face in RNN also



- suppose text generation problem where we need to predict the next word, if the short sentence is there then the short term dependency will be there which is ok but when a long sentence is there and we are trying to predict the next word then long term dependency will occur which cannot be captured by RNN leading to poor accuracy because of dependency
- above we have seen the weight updation till 3-4 length sentences but as the length of sentences increases:



- as we apply the chain rule and when the chain rule is very big, will be getting the t=1 value approx equal to 0 which will not lead any change to weight change so W(old)~W(new) leading to vanishing gradient problem
- remember the nearest word chain rule will never become 0 means they play important role in predicting the next word
- we can say that when the chain rule is big means that word will not play an important role in predicting the next word as it will lead to long term dependency



- so RNN is having the dependency to nearest word and not on farest words present in the entire sentence
- we can also try to change the activation function from sigmoid to tanh but this also will not imporve the much accuracy
- so in order to solve this problem:
    - use the relu or it's varient activation function
    - LSTM RNN -> Long Short Term Memory RNN
    - GRU RNN
