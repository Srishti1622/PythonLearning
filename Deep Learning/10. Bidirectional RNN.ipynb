### Types of RNN
- One to One RNN: one input, one output
- One to Many RNN: one input, many output
    - Ex: Image Captioning means image as input and get description of image in words
- Many to One RNN: many input, one output
    - Ex: Image Search means you provide a sentence and get image as an output
- Many to Many RNN: many input, many output
    - Ex: Language Translator



## Bidirectional RNN


- As we know, in simple RNN, LSTM RNN or GRU RNN we will be having the context of previous words so problems where previous context is enough to predict the next upcoming output or word, we can use those RNN
- But problems where we need to have to context of previous as well as upcoming words means we want to predict the output or word in between then will go with bidirectional RNN



- will have two RNN's whose input and output of each state will be connected together 
- in one RNN as forward propogation, will pass word by word 
- in second RNN as backward propogation, will pass the word by word buut in reverse direction of sentence
- suppose the sentence is "Srishti eats _______ in Bangalore" then:
    - in forward propogation, will pass Srishti as x11, eats aas x12 and so on
    - in backward propogation, will pass Bangalore as x11 and that to from backward direction 
- so the advance of this is that our model will have previous as well as upcoming context
