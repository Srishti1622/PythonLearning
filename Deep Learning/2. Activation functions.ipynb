## Activation Function
Activation functions helps to determine the output of a neural network. These types of functions are attached to each neuron in the network, and determines whether it should be activated or nooot, based on whether each neuron's input is relevant for the model's prediction. Activation function also helps to normalize the output of eah neuron to a range between 0 to 1 or -1 to +1.



#### Types of Activation Function
- Sigmoid Function
- Tanh Function
- ReLU Function
- Leaky ReLU Function
- Parametric ReLU Function

NOTE: z value means taking about forward propogation formula and output while derivative of z value means taking about backward propogation and output

### Sigmoid Function
Advantages:
- smooth gradient, preventing "jumps" in output values
- output values bound between 0 to 1, normalizing the output of each neuron
- suitable for binary classification problem
- clear predictions, ie., very close to 0 or 1

Disadvantages:
- It is responsible for transforming the z value between 0 to 1 and derivative of sigmoid will always ranging between 0 to 0.25 because of which **"Vanishing Gradient problem or gradient saturation or gradient dispersion"** occurs
- as we know that the threshold value is 0.5 so the function output is not centered on 0 (not zero-centered), which will reduce the efficiency of weights update - Standarization is process of making zero-centered dataset where mean is 0 (StandardScaler)
- performs exponential operations which is slower for computers are relatively time consuming



### Tanh Function
- it is responsible to transform the value between -1 to +1 and derivative of tanh will always ranges between 0 to 1 so till medium size problem will not face vanishing gradient problem but for deep size problem having lot of layers will still face the vanishing gradient problem
- this function is zero-centered means weight updation happens efficiently, which is better than sigmoid function
- it is also relatively time consuming as here also we have exponential operations
- In general binary classification problem, the tanh function is used for the hidden layers and sigmoid function is used for the output layer. however these are not static it will depends on specific problems



### ReLU (Rectified Linear Unit) Function
- the output value will be either 0 or +ve as max(0,z) whereas derivative of relu is always either 0 or 1 so it is solving the vanishing gradient problem as if derivative of relu output is 1 (z=+ve) means weight updation will happend and if 0 (z=-ve) means it will create a "dead neuron" - a neuron without any functionality
- calculation is super fast as reLU function has a linear relationship curve
- as the output of reLU function is max(0,z) which means it is not zero-centered



### Leaky ReLU Function and Parametric ReLU Function
- Leaky ReLU = max(0.01z , z), Parametric ReLU = max(alpha z , z) where alpha will be hyperparameter which we will be finding out using hyperparameter tuning
- so the derivative of leaky relu will never be 0, it will always be small +ve value and for values greater than 0 it will be 1
- Leaky ReLU has all the advantages of ReLU and here there is no "Dead Neuron" problem
