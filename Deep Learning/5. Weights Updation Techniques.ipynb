## Exploding Gradient Problem
- this problem happens because of weights initialization techniques where the weights initalized value is very big
- similar to vanishing gradient problem where because of derivative of activation function becoming very small leading to vanishing problem
- both the problem occurs while splitting the chain rule of derivative
- so here when the weights are very big value then either W(new)>>>>>>W(old)  or  W(new)<<<<<<<< W(old) which will lead to convergence went out of the curve and never reaches to global minima



## Weight Initialization Techniques
- to avoid facing exploding gradient problem, will try to use weights initialization techniques to initial the weights keeping in mind the key points:
    - weights should be small
    - weights should not be same
    - weights should have good variance

### Types of Weight Initialization Techniques
- Uniform Distribution
- Xavier/Glorot Initialization
- Kaiming He Initialization

#### Uniform Distribution
- whenever we initial weights that is based on some uniform distribution having range between a and b 
- Notation: Wij where i is previous layer and j is next layer



#### Xavier/Glorot Initialization
- invented by researcher named Xavier Glorot
- two types:
    - Xavier Normal Initialization: mean=0 and standard deviation=sigma
    - Xavier Uniform: uniform distribution from range between a and b



#### Kaiming He Initialization
- two types:
    - He Normal: all the weights will be taken from normal distribution
    - He Uniform: all the weights will be taken from uniform distribution range from a to b



## DropOut Layer
- as we known the concept of overfitting in which the model having very good training accuracy 90% but the test accuracy is not good 60%
- this overfitting problem may also occur in ANN when we have so many neurons in each layer specially in multi layer neuron netwrok and all those are connected to next layer neurons having so many weights and bias
- Like in machine learning, decision tree and random forest to reduce the overfitting problem to create a generalized model will do something like pre-post puring or feature and row sampling 



- so to reduce the overfitting problem in ANN, we used something called "DropOut Layer" to create a generalized model
- the dropout value ranges between 0<=p<=1 where p denotes the probability of how many neurons in that particular layer will be deactivated, we do hyperparameter tuning to select the p value
- suppose for input layer having 4 neuron and p=0.5 then 2 means 50% of neurons will be deactivated and for every epoch or iteration random neuron gets deactivated everytime based on probability value suppose in 1st iteration x2 and x4 was deactivated so in 2nd iteration x1 and x2 were deactivated and so on

NOTE: will apply dropout while training the model and deactivate the neurons based on probability but for testing the model with test data we don't deactivate any neuron instead will multiply the weights with probability for that particular layers
