## Exploding Gradient Problem
- this problem happens because of weights initialization techniques where the weights initalized value is very big
- similar to vanishing gradient problem where because of derivative of activation function becoming very small leading to vanishing problem
- both the problem occurs while splitting the chain rule of derivative
- so here when the weights are very big value then either W(new)>>>>>>W(old)  or  W(new)<<<<<<<< W(old) which will lead to convergence went out of the curve and never reaches to global minima



## Weight Initialization Techniques
- to avoid facing exploding gradient problem, will try to use weights initialization techniques to initial the weights keeping in mind the key points:
    - weights should be small
    - weights should not be same
    - weights should have good variance

### Types of Weight Initialization Techniques
- Uniform Distribution
- Xavier/Glorot Initialization
- Kaiming He Initialization
