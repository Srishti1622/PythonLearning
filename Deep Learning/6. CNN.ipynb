## Convolutional Neural Network
- as we know with the help of ANN will be able to solve supervised learning problems
- in CNN, the input data is basically **"Images"** and we try the solve the problem like image classification, object detection, segmentation
- so the part which we need to focus on for CNN which replicate part1(Occipitial Lobe) and part12(Cerebral Cortex) of the brain



### Human Brain vs CNN


- As soon as the signal comes to cerebral cortex it goes to visual cortex which is a present in occipitial lobe
- visual cortex is the region of the brain that receives, integrates and processes visual information relayed from the retina. usually visual cortex region is divided into 5 different layers (V1-V5)
    - V1(Primary Visual Cortex): cerebral cortex passes the signal through V1 it preforms orientation of edges and lines
    - V2: responses from V1 will passes through V2, it will responsible to differentiate in color, complex patterns like shape, object orientation
        - V2 can send the signal back to V1 or pass the signal to V3, V4, V5
    - V3, V4, V5: complex process like object recognition will happen in these layers
    - and finally will be able to visualize the image in occipitial lobe region
- In CNN, researchers tried to replicate the same process of human brain

#### RGB Images and GrayScale Images
- images are different into pixels and each pixel value will be ranging between 0 to 255 where 0 means black and 255 means white
- in grayscale images, will be having only one channel while in rgb images will be having 3 channels red, green and blue
- dimension: rows (height) pixel * columns (width) pixel * channels



### Convolution Operation in CNN
- Step1: Whenever we get the pixel dataset and we need to start the concolution operation, we first need to normalize the pixel which means converting all the pixel range between 0 to 1 (to normalize we divide all the pixels by 255)
    - in CNN, the V1 layer work of human brain is done by "filters" that is a 3X3 matrix here, filters can be of different different sizes
    - the main aim of these filters is to pick out the information from the images pixels



- Step2: Convolution Operation -> applying and placing the filter (basically multiplying the filter values with image pixel values and adding up all)
    - for applying the filter will do **"stride: jump towards right, left or bottom"**, here stride=1
    - here the filter which is applied is **"vertical edge filter"** which is responsible to finout the edge



- Just to make you understand doing: Denormalize the data what we got after applying filters, so 0 will become 255 and lowest number become 0
- Formula to get the output dimension: n-f+1=output_size where n is the size of input, f is size of filters
- so here n=6, f=3, o=4 (given 6X6 and getting 4X4 means we are lossing some information so to avoid the lose of information we used technique called **"Padding"**)

NOTE: filters are already pre-defined usually in convolution neuron network we don't defines the value of filters we randomly initialize the value and later on through forward and backward propogation, these value will get updated, so the size and value inside the filters will not fixed. Also in  one input image we can appply as many filters as we want and will gets that many output images

- Step3: apply ReLU or it's variants operation [max(0,z)] in each pixel of each output image, why ReLU is getting applied because we need to update the filter values so to update filter value we need to do backward propogation 



### Padding in CNN
- as when we applying filters on some pixel data then some information is getting lost to prevent that we do padding so that i can get the same pixel size data in output after applying filters
- padding basically means adding cushion and we can add any number of padding depending upon our output data
- suppose image size in input is 6X6X1 and after adding padding of 1 layer from all the sides, the input size becomes 8X8X1 but remember original image size will stay same
    - after adding the padding the main question is what to fill in those pixels?
    - so there are two types of padding:    
        - Zero Padding: in which fill the padding pixels with zero
        - Neighbour Padding: fill the padding to same as it's neighbour
- to know how much padding we need to add so that the image size stay same as input in output using formula **n-f+2p+1=o** here n is input size, f is filter size, p is padding which we need to find out, o is output size same as input size



### Max Pooling, Min Pooling, Mean Pooling
- so after one convolution layer (applying filter on input and then applying ReLU on output this complete is called as one convolution layer and we can have any number of convolution layers)
- Max Pooling: so after applying ReLU whatever output we got will apply max pooling filter on that 
    - with the help of Max Pooling we are focuing more on getting all the important information irrespective of filter in convolution layer
    - Location Invariant: it says that wherever in the image important information is there max pooling will able to extract those
    - so to apply max pooling filter the process is same but here instead of multiply and adding the value will pick the max value of selected filtered box
    - Also here the stride means jump value will be the equal to size of max pooling filter which means that it will not repeat the filter on same pixel box



- Min Pooling: everything is same as max pooling the only thing is instead of taking the max pixel of box will take the minimun value, it is used to focus on minimun intensity part of image
- Mean Pooling: here will take the average of selected filter box values
