## Optimizers
Optimizers are basically responsible for reducing the loss function in the backward propogation by updating all the weights (with weight updation formula)
- Gradient Descent
- Stochastic Gradient Descent (SGD)
- Mini Batch with SGD
- SGD with Momentum
- Adagard
- RMSPROP
- Adam Optimizer

### Gradient Descent Optimizer
- when while plotting graph of weights with respect to loss or cost function, if we get parabola curve which is known as gradient descent curve in which our main aim is to reach the global minima where slope is 0
- Two important terms related to gradient descent optimizers specifically for cost function:
    - Epochs: 1 epoch means one forward propogation with all the datapoints and one backward propogation that will update the weights, so we may have n number of epoch until the cost function gets reduced 
        - An epoch refers to one complete pass of the entire training dataset through the model. When training a neural network, you typically run multiple epochs to ensure the model learns effectively
        - Example: If you have 10,000 training samples and train the model for 5 epochs, it means the model has seen the entire dataset 5 times.
    - Iteration: 1 epoch === 1 iteration
        - An iteration is a single update step of the models parameters
        - It depends on batch size (how many samples are processed in one step). One iteration occurs every time the optimizer updates the models weights.
        - Example: If you have 10,000 training samples and use a batch size of 100, then: 
            - You need 100 iterations to complete one epoch (since 10,000  100 = 100).
            - If you train for 5 epochs, the total number of iterations will be 100  5 = 500.

    

Advantages:
- Convergence will happen

Disadvantages:
- Resource Intensive: Require huge resource like RAM, GPU as for each epoch, forward and backward propogation here we are taking all the data points everytime

### Stochastic Gradient Descent (SGD)
- here in 1 epoch will be having total iteration equal to number of data points present in the dataset
- suppose if we have 1000 datapoints then 1 epoch will have 100 iterations
- and the number od epoch we can have depends on loss function so until the loss function is max reduced



Advantages:
- Solve resource intensive issue of gradient descent optimizer

Disadvantages:
- Time complexity will be huge as here we are taking 1 datapoint at a time
- as the time complexity is more means convergence will also take time 
- Noise gets introduced as convergence is not smooth, here noise means the path to global minima will have lot of zig-zag and not smooth going

### Mini Batch SGD
- here will have a new term called "batch_size"
- so here instead of taking 1 data point in each iteration will be taking batch size data points 
- so number of iterations in 1 epoch = number of data points / batch size



Advantages:
- Convergence speed will increase as comapred to SGD
- Noise will be less as compared to SGD
- Efficient resource usage like RAM

Disadvantages:
- some amount of noise will still exists but less than SGD
