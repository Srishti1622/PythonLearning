## Loss Function vs Cost Function
- Our main aim is to reduce the loss function or cost function with the help of optimizers like gradient decent by updating the weights in backward propogation
- whenever we talk about loss function, will take every single data point and do forward and backward propogation, so if we want to play with every data point and update the weights then use loss function
- whereas in cost function, instead of taking one point we take all the points together or bunch of points and all the weights gets updatedat once



### Regression Cost or Loss Function
- Mean Squared Error (MSE): if don't have outliers in dataset then use this, even we can remove outliers and then use this but removing outliers will lead to removing some amount of data 
- Mean Absolute Error (MAE): if have outliers in dataset then use this
- Huber Loss: combination of MSE and MAE
- Root Mean Squared Error (RMSE)

#### Mean Squared Error (MSE)


Advantages:
- MSE is differentiable means at any point in curve of gradient decent will be able to find out the derivative change or weight updation
- It has 1 local or global minima only
- It converges faster as we are getting smooth curve because of quadratic equation

Disadvantages:
- Not robust to outliers as because of outlier we need to penalizing (means squaring) the cost function with respect to error which lead to major angle movement of best fit line towards the outlier

#### Mean Absolute Error (MAE)


Advantages:
- robust to outliers as here we are not penalizing the error

Disadvantages:
- convergenes usually takes time as here we don't get parabola curve so we need to take sub gradients to calculate the slope

#### Binary CROSS Entropy


- here the loss function is based on logistic regression log loss function

#### Categorical CROSS Entropy


- in this, first One Hot Encoding is getting applied in the output feature



- here as predicted ouput will get the probability of each category for that data point like what percentage each category is belong based on data point

#### Sparse Categorical CROSS Entropy


- the main difference between categorical cross entropy and sparse categorical cross entropy is here we get the index of category as final output instead of getting the probability of each category
- it will just provide the final output category based on the highest probability of category
- disadvantages is here we are lossing the information about the probability of other categories
