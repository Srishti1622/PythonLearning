## Deep Learning
- ANN (Artificial Neural Network): will able to solve classification and regression problem
- CNN (Convolutional Neural Network): here the input data will be in the form of images and video frames
    - RCNN, Masked RCNN -> Object Detection -> Computer Vision
- RNN (Recurrent Neural Network): usecase is related to NLP, input data will be in the form of text or time series 
    - Word Embedding, LSTM RNN, and so on
