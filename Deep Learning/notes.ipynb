## Preceptron (Artifical Neuron or Neural Network Unit or Single Layered Neural Network)
- Preceptron is a basic unit of ANN
- It is used to solve binary classification problem



- Terms in Preceptron:
    - Input Layer: The number of inputs in input layer will be the number of independent features in the dataset
    - Hidden Layer: In preceptron, will only have one neuron, and in this neuron will perform two task or precessing:
        - Step1: z = summation of  WiXi + b where b is bias, this bias is treated as a noise which plays an important role

        
        
        - Step2: apply activation function on top of z to transform the output value between some value like 0 to 1 or -1 to +1
    - Weights: the input layer is connected to hidden layer will have some weight. Weights plays a very important role in training a good accuracy model
    - Activation Function: the main aim of applying activation function is as we are solving the binary classification problem ao after performing the process will be getting either 0 or 1. there are many activation function with some threshold values like Step function, Sigmoid function

    

    - after performing the processing **"Feed Forward Neural Network"** suppose we got output value and if the predicted output and original output is same which means error is 0 hence the weights whatever we have applied is good for the model
    - but suppose if the error is high, we need to update the weights and this process will go on untill we get weights that satified the original output

### Advantages and Disadvantages of Preceptron
- It is just good for solving Binary Classification Linearly Separable Problem
- As going again and again and update the weights everytime when we got errors is not efficient technique
- to solve the problem which are not linearly separable , we need to apply and use Multi Layered Neural Networks

## Multi Layered Preceptron Model (Aritifical Neural Network)


- In multi layered neural network, we can have more than one, any number of hidden layer and neurons and each neuron will have some constant bias value and also that 2 step process will also happen for every neuron
- so whatever neuron will be last will be treated as output layer, based on the claasification problem type will have number of neurons in output layer like if binary classification then will have one neuron, if trinary classification then will have two neurons and so on
- our main aim is to reduce the error so that our model get trained properly
- terms to remember:
    - **forward propogation:** whatever process we do in feed forward neural network will perform all the process here

    
    
    - **backward propogation:** whenever we get high error, we need to reduce the error by performing backward propogation, we try to update every weights from last to first (right to left <--), and when all the weights get updated will try the model with next input and find out the predicted output and error, so this entire process will happen until we reduce our loss function or error

    

    - **loss function:** this loss function will tell how much error we got, so whatever final predicted output we will be getting from ANN, error = actual output - predicted output, to update the weights will use **"Weight Updation formula"** and plot a graph of weights with respect to loss function used and get the gradient decent

    

    - **optimizers:** to reduce the loss function or error we use optimizer called as gradient decent optimizer, and aim is to reach global minima which is when W(new) == W(old). There are different types of optimmizers. The main aim of optimizers is to perform backward propogation and update weights. 

    

    - **chain rule of derivative:** splitting the derivative based on dependency of output and weights

    

    

    NOTE: whenever we apply sigmoid activation function on z value will going to get the value between 0 to 1 but when we apply sigmoid function on derivative of z value then always will going to get the value between 0 to 0.25 (mathematically proven)

    

    - **vanishing gradient problem because of sigmoid activation function:** as the derivative value of sigmoid function is always between 0 to 0.25 which is a small value and every time will multiplying those value for each in chain rule will result in even more smaller value which lead to approx no change in weight update value, so convergence never happens as it will not move towards the global minima. To fix the sigmoid vanishing gradient problem, will use other activation functions
