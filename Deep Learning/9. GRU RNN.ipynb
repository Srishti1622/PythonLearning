## GRU RNN - Gated Recurrent Unit
- introduced by Cho,etal in 2014



- as we have seen LSTM RNN having a complex architecture like three different gates and large trainable parameters like bias and weights, so as the input size increases or neurons in the hidden layer increases, the number of trainable parameters will also increase and will consume lots of time in training the model
- so in GRU RNN, instead of using two different long term memory and short term memory separately, we have combined both the long term and short term memory cells 
- Instead of having forget gate, input gate and output gate, here we will be having "Update Gate - z(t)", "Reset Gate - r(t)" and "Temporarycandidate Hidden Gate - h'(t)"

##### Reset Gate
- It is responsible for reseting some information from h(t-1) which is nothing but memory (long term and short term) then we apply pointwise multiple operation in h(t-1) and r(t) which will reset the value



##### Update Gate
- It says that what context information needs to be addeed but this will be completely dependent on candidate hidden state which talks about current context. If current context is important then add the information with less information from z(t) and more information from current context and vice versa
