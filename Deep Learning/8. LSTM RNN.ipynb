### RNN vs LSTM RNN


- as we know in RNN we have feedback loop which is basically for short term memory
- in lstm, as RNN is there so we already having short term memory but here we also have long term memory which is like a ongooing straight line known as "Memory Cell" in which we can store the context which is important and remove the context at any point (conside this long term memory line as luggage belt in airport)

## LSTM Architecture
- It has 3 important gates:
    - Forget gate
    - Input gate and Candidate memory
    - Output gate

Understand the important Notation:



### Forget Gate
- in forget gate, we are combining/concatenate the vectors of h(t-1) and x(t) and then pass the result of concatenate to hidden layer neurons with sigmoid activation function, as we known signoid function will give output in range from 0-1 so F(t) will be containing vector of 0's and 1's
- then the F(t) vector will go and apply point multiple operation with vector coming from memory cell
- in forget gate, as name suggests will remove or forget some context or some part of context or keep the context same 





### Input Gate and Candidate Memmory
- the way we calculated the value of F(t) in forget gate, same way we will calculate the value of I(t) 
- for calculating candidate memory, the process is same but here instead of applying sigmoid activation function will apply tanh activation function
- then we concatenate the outputs I(t) and C'(t) to get our final input gate result
- then this final output of input gate and the vector coming from memory cell will perform the point add operation which will lead to add new context to memory cell and result in final C(t)
