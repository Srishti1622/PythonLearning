## Silhoutte Clustering
It helps us to validate the clustering algorithms.

There are some steps which we need to follow to validate the clustering algorithm:
- Step1: We have taken one cluster, in that cluster take one point and from that point calculate the average distance of sum of the points from that selected point


- Step2: then define the mean dissimilarity -> take the nearest cluster (which will also have some points) present to the above selected cluster, now calculate the average summation of distance from the above selected points to the points present in the nearest cluster

NOTE: so here we can notice that if the clustering is done well, then a(i) << b(i)


- Step3: after finding out the a(i) and b(i), now calcualte the Silhuotte Score
    - silhoutte score will ranges from -1 to +1
    - more near to +1, better the clustering model formed
    - more near to -1, clusters are not good
