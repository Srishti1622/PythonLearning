## K Nearest Neightbour (KNN) Machine Learning Algorithm
- It can be used to solve both "Classification" and "Regression" problems
- It is the most easy to learn

### KNN Classification


Steps to perform KNN on classification problem:
- We have to initialize the K value, K can be any number greater than 0 and can go upto infinity
- We can use hyperparameter tuning to find out the best K value based on dataset
- Find the K nearest neightbour for the new test data point
- From those K values, how many neighbours are belong to each category
