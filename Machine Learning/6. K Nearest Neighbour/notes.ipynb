## K Nearest Neightbour (KNN) Machine Learning Algorithm
- It can be used to solve both "Classification" and "Regression" problems
- It is the most easy to learn

### KNN Classification


Steps to perform KNN on classification problem:
- We have to initialize the K value, K can be any number greater than 0 and can go upto infinity
- We can use hyperparameter tuning to find out the best K value based on dataset
- Find the K nearest neightbour for the new test data point
- From those K values, how many neighbours are belong to each category

#### How to find the K nearest neighbour
To find out the K nearest neighbour, we use their distance, so there are two distance formula which will be using to find out the K nearest neighbours:

Both are used to find out the distance between two points:
- **Euclidian Distance:** flights between two places uses this method to calculate the distance


- **Manhattan Distance:** uber cab, road mediums uses this method to calculate the distance



## KNN Regression

- all the steps are same as claasification
- here after finding out the K nearest neighbours, then find out average or median of those all points and that will be the final result for new test point

## Variants of KNN
- There is one problem with KNN that if we need to find out the K nearest neighbour, will be measuring the distance of that new test point with all the training points present which will increase the time complexity 
- so to decrease the time complexity for which our aim should be that we should not go and find out the distance of each points for which will having two variants of KNN:
    - **KD Tree**
    - **Ball Tree**

### KD Tree

- It is in a form of **"Binary Tree"** and Backtracking is also used to find out the K nearest neighbours
- Steps:
    - Will take the median of each independent feature
    - using that median, will try to split it in multiple parts
