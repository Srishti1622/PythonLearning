## Naive Baye's Algorithm
- This algorithm is specifically for solving **"Classification"** problems (both binary and multiclass)
- To understand this algorithm, we must need to have an understanding of **"probability"** and **"baye's theorem"**

### Probability
- Two types of events occurs:
    - Independent events:
        - Rolling a dice: probability of one outcome is not changing the probability of other outcome

        
    - Dependent events:
        - probability of one outcome will change the probability of other outcome

        

### Baye's Theorem
- Baye's theorem is based on the dependent event of probability



- P(A/B): probability of event A given B has occured
- P(A): probability of event A 
- P(B): probability of event B
- P(B/A): probability of event B given A has occured

### How will use Baye's theorem in machine learning
- Suppose we have x1, x2, x3 are independent features and y as dependent feature having binary classification of yes/no


- so will try to solve this using baye's theorem


- after getting the probability of each category, final result will be the category whose probability is more

NOTE: **the algorithm that specifically use baye's theorem is called as Naive Baye's Algorithm**

## Variants of Naive Baye's
- Bernoulli Navie Bayes
- Multinomial Naive Bayes
- Gaussian Naive Bayes

NOTE: Variants are important as based on different types of datasets will need to findout which navie bayes algorithm suits best

### Bernoulli Navie Bayes
- Whenever independent features are following bernoulli distrubution, then we need to use bernoulli naive bayes algorithm
- bernoulli distribution basically means when the value of independent features is either 0 or 1 (like tossing a coin, success/failure, pass/fail, male/female)
- and output feature may be either binary classification or multiclass classification

### Multinomial Navie Bayes
- Whenever we have independent or input features in the form of text, we can use multinomial naive bayes algorithm
- like spam filter classification
- so to convert text to vectors, will use NLP techniques as machine learning model doesn't understand text

### Gaussian Navie Bayes
- If the independent features are following gaussion distribution, then we need to use gaussian navie bayes algorithm
- Gaussion distribution basically means **Bell Curve** like normal distrubtion or (little left or right skewed) or convert any other distribution to normal
- here the dataset will have continuous values in independent features like age, height, weight and we need to predict whether the person is over-weight or not

NOTE: there might be a case where in a dataset, will have some features as bernoulli and some as gaussion, then we need to check most of features follows what and choose that navie bayes algorithm
