before understanding PCA, we need to first understand why we need PCA (Principal Component Analysis - Dimensionality Reduction)

## Curse of Dimensionality
- dimension means number of input features
- Suppose we have 6 models and 500 features, for each model we have some accuracy based on the number of features they are using 
- so when we are providing only the important and less features to the model, the accuracy was high but as soon as we start increasing the feature size like 50, 100, 500, the model accuracy start decreasing as not all features are as important as others for prediction
- some feature may be very important, some less important, some not at all important
- so we can say that the model is overfeeded, the model should be given with that number of features till which the accuracy keeps on increasing
- **as the number of features increasing, the model will try to learn all of those, leads to confusion which leads model performance degrade**

### Different ways to prevent Curse of Dimensionality
There are two ways to remove Curse of Dimensionality:
- **Feature Selection:** will take only important features and train the model
- **Dimensionality Reduction -> Feature Extraction -> PCA:** will try to derive a feature from the set of features capturing much essence of previous features



### Why Dimensionality Reduction?
- To prevent the curse of dimensionality
- To improve the performance of the model
- To visualize the data -> as it will help us to understand the data, as human being we can't see more than 3d figures so it will help us to reduce the dimension even from 100d to 3d or 2d so that we can visualize it

### Feature Selection vs Feature Extraction
- Feature Selection is a process which helps us to select the most importamt feature which will help us to predict the output



- if linear relationship means when x increasing y increasing, when x decreasing y decreasing
- if inverse linear relationship means when x increasing y decaresing, when x deacresing y increasing



- Mathematical way to quantify the relationship between x and y is called as **"Covariance"**
    - if its a positive value then linear relationship
    - if negative value then inverse linear relaationship - and if its approx equal to 0 then no relationship in x and y
    - the covariance value may range from any positive to negative value



- **"Pearson Correlation"** its value ranges from -1 to +1
    - the more towards the value of +1 the more positively correlated x and y is
    - the more towards the value of -1 the more negatively correlated x and y is
    - the value approx equal to 0 means no relationship



- Feature Extraction will reduce the dimension of features by performing dimensionality reduction 
- Suppose there are two features and both are very important to predict the output so we can't perform feature selection and rop the any feature, in this case we apply some transformation to extract new feature



## PCA (Principal Component Analysis)
- used for dimensionality reduction - feature extraction process



- suppose we have two features, size of house and no. of rooms and we want to predict the price before that we want to convert the 2d to 1d
    - so using feature selection, we can simply project the data points in feature
    - the area from the first data point to last data point will be called as **"Spread"**
    - if the spread is huge, the variance will increases
    - disadvantage here is one feature is getting captured but the other important feature is getting lost which is also having huge spread
    - which will lead to poor model performance



- so to prevent the loss of information while doing feature extraction in PCA we do some transformation by applying some mathematical equations (Eigen decomposition on matrix)
- it helps in finding out the best Principal Component line which captures maximum variance
- will be getting new axis and will try to prject the data points in that new axis
- and other axis will be perpendicular to the second axis
- so one axis will have hude spread while the other does not leading less information loss

NOTE: number of new axis created suppose three, then PC1, PC2, PC3 named based on the high variance to low variance as var(PC1)>var(PC2)>var(PC3)
