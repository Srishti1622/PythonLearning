before understanding PCA, we need to first understand why we need PCA (Principal Component Analysis - Dimensionality Reduction)

## Curse of Dimensionality
- dimension means number of input features
- Suppose we have 6 models and 500 features, for each model we have some accuracy based on the number of features they are using 
- so when we are providing only the important and less features to the model, the accuracy was high but as soon as we start increasing the feature size like 50, 100, 500, the model accuracy start decreasing as not all features are as important as others for prediction
- some feature may be very important, some less important, some not at all important
- so we can say that the model is overfeeded, the model should be given with that number of features till which the accuracy keeps on increasing
- **as the number of features increasing, the model will try to learn all of those, leads to confusion which leads model performance degrade**

### Different ways to prevent Curse of Dimensionality
There are two ways to remove Curse of Dimensionality:
- **Feature Selection:** will take only important features and train the model
- **Dimensionality Reduction -> Feature Extraction -> PCA:** will try to derive a feature from the set of features capturing much essence of previous features
