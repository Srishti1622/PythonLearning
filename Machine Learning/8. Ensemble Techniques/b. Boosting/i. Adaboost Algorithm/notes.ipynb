## AdaBoost Machine Learning Algorithm

- here all the weak learner model are **"Decision Tree"**
- all the models are connected in sequence order
- so whatever records are wrongly predicted by one model will give those records to other next model
- and final we will get strong learner

NOTE: Weak learners means they haven't learnt much from the traning dataset 

IMP: so in random forest, for classification, we have used majority voting and in regression, average of all the model will be as the final output but **in Adaboost, we assign some weight to weak learners, the weights may be higher or lower**



here M1.....Mn are **"Decision Tree Stumps"** and α1.....αn are **"weights"** assigned to weak learner models

### Decision Tree Stumps
"Decision Tree Stumps" means the depth of decision tree will be 1, so we call it as weak learners, and it will leads to **"underfitting"** problem in which training accuracy is less, so when we combine all the weak learners decision tree stumps will leads to strong learner



## AdaBoost Classifier Maths Indepth Intuition Steps

### Step1: Create Decision Tree Stumps
- we will create decision tree with each feature
- then using entropy, gini impurity and information gain, will find out which feature is best suited for decision tree stump



### Step2: Sum of Total Errors and Performance of Stump
- To find out the total error, we need to first assign some sample weights to all the records
- Count the total number of records and provide equal sample weights to all the records as **"1/number of records"**
- Find out the wrong records present as error
- Calculate the total error, **Total Error = Sum of all the sample weights of wrong records**
- Calculate Performace of Stump using formula mentioned in the image
- Now the calculated performance of stump will be the actual weight means α1 (alpha1 as this is the case of M1 weak learner)


- in the mentioned example, when the credit score is good then there is no records with No as output which means the 1yes in No output is an error

### Step3: Update the weights for correctly and incorrectly classified points
- Now our aim is that we will send the wrong records to the next decision tree stump
- So for this, will assign less weights to correctly classified points and high weights to incorrectly classified points
- This will increase the probability of selecting the wrong records by the next decision tree stump
- **For correctly classified points = old weight X e^(-performance of stump)**
- **For incorrectly classified points = old weight X e^(performance of stump)**



### Step4: Normalize Weights Computation and Assigning Bins
- If we do the sum of all the updated weights, it is not equal to 1
- Sum the updated weights
- To normalize, divide each updated weight with the sum of all updated weights
- now for next model data point selection, will assign bins to each records so that the chance of selecting wrong records increases
- In Bins Assignment, will create a range of bins for each records, first records have bins range of 0-normalized weights, after that just add normalized weights and assign bins range (keep on adding the normalized weights)
- We can notice that the bins range size of wrong records will be max/high as compare to correct points



### Step5: Select data points to send to next Stump
- We will doing an iterative process in which will be selecting random values between to 0 and 1
- And based on that random value, will choose the respective bins record as data point will want to send
- So as the bin range of wrong records is max, there are more chances that the wrong records get selected more 



### Step6: Send the selected records for training to next Stump
- again the same process from step1 till here will start 
- in each iteration, will get alpha value means weight value for the weak learner model

### Step7: Final Prediction
- Now new data point is there for prediction
- We will get the output of each weak learner model for that new data point
- will do calculation, multiple each alpha weights with respective model output
- atlast, whichever output is having more weight will be the final output for the new data point





## AdaBoost Regression
- All the steps will stay as same as in Adaboost Classifier
- The only difference will be:
    - for selecting best feature: in classifier we used entropy, gini impurity and information gain but in regression, will use variance and variance reduction
    - the output of each model will be contiunous values
