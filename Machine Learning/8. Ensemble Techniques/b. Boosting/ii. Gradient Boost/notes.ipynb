## Gradient Boosting Algorithm
- it is used to solve both Classification and Regression problem 
- it is a part of boosting ensemble technique where we create decision tree sequentially, we combine all the weak learners and final get the strong learner



### Gradient Boosting Regression
Steps:
- **Create a base model:** make sure the base model is not baised to any value, it will be giving you some default value, to create a base model will find out the average of output feature which will be the default value for base model
- **Compute the residuals or error:** to find out the errors for each record, subtract the actual output with base model default value basically (actual output ,y and predicted value, )
- **Construct a Decision Tree:** Consider input features xi and output as Ri (residual) instead of actual output y, will getting residual R2
- **Compute new predicted output:** it will leads the overfitting suitation, so to avoid overfitting, will use **Learning rate**
- The value of Learning Rate ranges from 0 to 1
- so **predicted output = base model random value + (Learning rate * Decision Tree output)**
- after new predicted output, again find out the new residual error values and so on
