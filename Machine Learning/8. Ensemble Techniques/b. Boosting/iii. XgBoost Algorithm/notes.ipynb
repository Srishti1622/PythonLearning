## Xgboost Algorithm
- it is used to solve both Classification and Regression problem

### Xgboost Classification
Steps:
- Construct a base model, as in binary classification the probability will be 0.5, base model will not be baised to any output value
- Construct a Decision Tree with root, so before constructing the decision tree, we need to find out the residual1 which will be actual output values - base model value
- so for constructing decision tree, will take input features and residual1 as output feature


- Calculate Similarity Weight for each node, here residual square means sum of residuals square, probability will be of base model and need to take it as number of values present in that node
- Calculate Gain, similarity weight of left child + similarity weight of right child - (similarity weight of root node) 
- above similarity weight and gain are calculated to get the best feature to perform split for decision tree
- so which spilt feature will having higher gain value, will going to take that as best feature for split in decision tree
