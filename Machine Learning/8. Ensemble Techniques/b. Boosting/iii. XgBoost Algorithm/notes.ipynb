## Xgboost Algorithm
- it is used to solve both Classification and Regression problem

### Xgboost Classification


Steps:
- **Construct a base model**, as in binary classification the probability will be 0.5, base model will not be baised to any output value
- **Construct a Decision Tree with root**, so before constructing the decision tree, we need to find out the residual1 which will be actual output values - base model value
- so for constructing decision tree, will take input features and residual1 as output feature


- **Calculate Similarity Weight for each node**, here residual square means sum of residuals square, probability will be of base model and need to take it as number of values present in that node for each residual, here in formula DT1, DT2 all are similarity weights for that tree
- **Calculate Gain**, Gain = similarity weight of left child + similarity weight of right child - (similarity weight of root node) 
- above similarity weight and gain are calculated to get the best feature to perform split for decision tree
- so which spilt feature will having higher gain value, will going to take that as best feature for split in decision tree
- all the steps will repeat untill will get strong learner

NOTE: in the base model, whenever a new test data comes, it has to pass through log(odds) formula, as can't directly use probability



#### Final Output for new records
- sigmoid activation always give value between 0 to 1, hence will going to get a good probability



NOTE: the question is untill what depth we need to construct each decision tree, so when demominator of smilarity weights is greater then the similarity weight then we can stop, **this demominator value is called as "Cover Value"**
