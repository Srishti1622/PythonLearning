## Random Forest Classification and Regression

- here all the models of base learners are **"Decision Tree"**
- to give sample of data to each model, will going to do **"Row Sampling and Feature Sampling"** means picking up some rows and features, and providing it to models
- In case of classification, the final output of new test data point will be the majority of output we got from all models
- In case of regression, the final output will be the average of output of all the models
- the major disadvantage here is the time complexity to train this model will be very high but accuracy will be better as so many base learner are involved

## Why should we use Random Forest instead of Decision Tree?

- if we create a decision tree using by default parameter values, it will cause overfitting means train accuracy high(low bais), test accuracy low(high variance)
- so our aim is to convert high variance to low variance with the help of random forest, since it have multiple decision tree with different set of dataset which will increase the accuracy
