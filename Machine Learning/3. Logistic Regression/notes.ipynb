## Logistic Regression
- used to solve the binary classification problem
- in binary classification, dataset will have output feature as binary categories like yes/no, pass/fail and so on
- here we can **"squash the line** so that will always get the point values between 0 to 1 in case of binary classification

### Why can't we use linear regression to solve binary classification problem

- in case of outlies, best fit line changes
- will get value >1 and <0 as squashing line is not possible

### How Logistic Regression solves Classification problem
- IMP: will apply **"Sigmoid Activation Function"** on top of best fit line to squash the best fit line
- Always the points in sigmoid graph lies betwwen 0 to 1 and always when z>=0 then sigma(z)>=0.5




## Cost Fucntion
- here in logistic regression, cost function will not give "gradient descent" as it will give "non-convex function"



in order to prevent the issue of stucking in local minima in case of non-convex function using "Log Loss" so that will get a convex function and will able to get "gradient descent": our aim is to always minimize the cost function by changing the values of coefficients and intercept




## Performance Metrics of CLassification problems
- List of metrics used to measure the performance of logistic regression classification models:
    - Confusion Matrix
    - Accuracy
    - Precision
    - Recall
    - F-Beta Score

- once predicting the output, first thing will do is to create a "confusion matrix"

### Confusion Matrix
- in case of binary classification, it's a 2x2 matrix
- always, diagonal values of the matrix are the actual correct values
- other than diagonal values are actually the incorrect or wrong results predicted by the model
- IMP Notations:
    - when actual is 1 and predicted is also 1, then True Positive (TP)
    - when actual is 0 and predicted is also 0, then True Negative (TN)
    - when actual is 0 and predicted is 1, then False Positive (FP)
    - when actual is 1 and predicted is 0, then False Negative (FN)
