## Support Vector Machines ML Algorithm
- helps us to solve both "Regression" and "Classification" problems
- In case of Regression: **"Support Vector Regression"**
- In case of Classification: **"Support Vector Classifier"**

### Support Vector Classifier (SVC)
- In SVC, along with best fit line, we also draw two marginal line (2d) or plane (3d)
- Both the marginal line should be in equi-distance and the distance is denoted by **"d"**
- The marginal line or plane is created in such a way that the distance between the planes should be "maximum"
- So the nearest point thourgh which the marginal line passes is called as **"Support Vectors"**



### Soft Margin and Hard Margin 
- Whenever there are some amount of error always present, we can't completely divide or make partition, will get few or lot of overlapping points, that is called as "Soft margin"


- When we are able to clearly separate the points and don't have any errors, that is called as "hard margin"

### Support Vector Machine (SVC) Maths Intuition
- as we know, the points below the best fit line making the angle more than 90, will be always "-ve"
- while the points above the best fit line making the angle less than 90, will be always "+ve"
- **"Unit Vector (magnitude of vector is 1):** to achiev this, we need to divide the equation (which is distance betwwen the two marginal lines) with ||w||
