```python
import pandas as pd
import json
from langchain_groq import ChatGroq

model=ChatGroq(model='llama-3.1-8b-instant')
# res=model.invoke("provide a dataframe in dict type so that i can directly store it in a variable of 10 records having key names as order_id,product,category,quantity,price,order_date. " \
# "Make sure that the dataframe should be in dict type. Also make sure you only provide dataframe without any additional information." \
# )
res=model.invoke("""
            You are an expect in creating dataframe having some duplicates keeping in mind the pandas dataframe rules. 
            I want you to provide the output in JSON format having order_id,product,category,quantity,price,order_date keys with data for dataframe.
            Make sure to only provide json as an output, don't include any addtional information or text.
        Instructions:
                 1. Ensure all lists for the keys have exactly 20 records.
                 2. make sure it will have duplicate records.
                 3. Only 'quantity' or 'price' column will have some missing values.
            Output format:
                 {
                 'order_id':[],
                 'product':[],
                 'category':[],
                 'quantity':[],
                 'price':[],
                 'order_date':[],
                 }
"""
)
df= json.loads(res.content)
df
```

```python
df=pd.DataFrame(df)
df.head()
```
