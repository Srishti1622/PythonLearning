## Why Containers?

#### Before using containers:



- Suppose in a team Developer-A install some libraries and packages to build an application and the application is running on his local machine
- Now if any other team member wants to run that application, they need to do the complete setup like install the libraries and packages
- So there might be an issue occur of mismatch libraries version which will cause application failure
- Similarily, suppose Dev env is ready now when QA team started testing in QA env they also faced same issue of library mismatch which leads to some module of application will not work
- Conclusion: for every new machine or server, we need to do the setup of whole application from scrtach by installing the required libraries and packages, also libraries version mismatch is big issue which will lead to application failure
- whenever we want to move or deploy our application in some other environment, we need to install all the dependencies
