## Why Containers?
- Containers is a way to package applications with all the necessary dependencies and configuration
- A Container is a **"Portable Artifact"** which means it is easy to share and move this package to any environment
- It makes development and deployment more efficient, easy in terms of making sure that all the required dependencies are there and in sync

#### Before using containers:



- Suppose in a team Developer-A install some libraries and packages to build an application and the application is running on his local machine
- Now if any other team member wants to run that application, they need to do the complete setup like install the libraries and packages
- So there might be an issue occur of mismatch libraries version which will cause application failure
- Similarily, suppose Dev env is ready now when QA team started testing in QA env they also faced same issue of library mismatch which leads to some module of application will not work
- Conclusion: for every new machine or server, we need to do the setup of whole application from scrtach by installing the required libraries and packages, also libraries version mismatch is big issue which will lead to application failure
- whenever we want to move or deploy our application in some other environment, we need to install all the dependencies

#### After using containers:
- Example:

    - Suppose someone is shifting house, from house1 to house2 so while shifting he will be taking all the required stuffs from house1 to house2 
    - So he first pack all the required stuffs in some container and then use some shipping transport to ship that container to house2
    - and unpack all the packed stuffs in house2


- same way the container work, we first pack all the required dependencies in one environment and then transfer that packed container to other environment and unpack the conatiner to use


## Docker
https://docs.docker.com/get-started/docker-overview/

Docker is an open platform for developing, shipping, and running applications. Docker enables you to separate your applications from your infrasturcture so you can deliver software quickly.

here applications means containers, basically docker is an platform which help us to create containers

### Docker Image and Containers

- To convert an application having dependencies and configuration into container, we need to containerzied or dockerized the entire application
- In terms of container, we create **"layers of images"**
- Whenever creating a container, whatever images used with respect to dependencies, if combine all the images together it is called as **"Docker Iamge"**. Docker Image are very small.


- So if we run the docker image, then internally it will first create a container basically an environment considering all the dependencies it will run the application 



### Docker and Virtual Machine
- Both task is to perform virtualization



Operating system: we have different layers, in different different operaing system version mostly the ui(application layer) changes, os kernel will stay almost same



- Docker will be virtualizing the Application Layer of the operating system that basically means docker images will be able to communicate with the OS kernels host
- VM actually virtualizes both the application layer and OS kernels layer, means whenever we install the VM, it will have it's own application layer and OS kernel. That means a VM is a complete separate sub-system which we can install in host OS
- Docker image size is usually smaller(MB) as it just virtualize the application layer but in case of VM, the size will be huge(GB)
- Docker containers start and run much faster compared to VM
- In terms of compatibilty, we can install VM in any OS but docker image there may be a compatibility issue like we can install Linus docker image on windows OS unless and util the OS support that as windows 10 and above supports linus
